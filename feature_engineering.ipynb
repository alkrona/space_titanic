```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data/train.csv')
df.head()
```

Here i want to drop PassengerID, Cabin , Name, 

```python
df.drop(['PassengerId','Cabin','Name'],axis=1,inplace=True)
df.tail()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.head()
```

```python
df.drop(['HomePlanet','Destination'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['CryoSleep']=df['CryoSleep'].map({False:0,True:1})
```

```python
df['VIP']=df['VIP'].map({False:0,True:1})
df.head()
```

```python
df['Transported']=df['Transported'].map({False:0,True:1})
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
X_train = df.drop('Transported',axis=1)
y_train = df['Transported']
y_train.tail()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [1,2,3,5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200,250,300]
}

from sklearn.model_selection import GridSearchCV

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
rf_best = grid_search.best_estimator_
rf_best
```

```python
%pip install joblib
```

```python
import joblib
filename = 'model/rf_model_1'
joblib.dump(rf_best,filename)
```

```python
imp_df = pd.DataFrame({
    "Varname": X_train.columns,
    "Imp": rf_best.feature_importances_
})

imp_df.sort_values(by="Imp", ascending=False)
```
